## whole, raw counts

```r
physeq.whole
```

```r
otu_table(physeq.whole)
```

```r
# convert data from phyloseq to deseq2 type
diagdds.w = phyloseq_to_deseq2(physeq.whole, ~ Material)
```

```r
# function to calculate geometric means
gm_mean = function(x, na.rm=TRUE){
  exp(sum(log(x[x > 0]), na.rm=na.rm) / length(x))
}
```

```r
# fit DESeq2 model
geoMeans.w = apply(counts(diagdds.w), 1, gm_mean)
diagdds.w = estimateSizeFactors(diagdds.w, geoMeans = geoMeans.w)
diagdds.w = DESeq(diagdds.w, fitType="local")
```

```r
# retrieve results
res.w = results(diagdds.w)
res.w = res.w[order(res.w$padj, na.last=NA), ]
```

```r
res.w
```

```r
# extract significant values
alpha = 0.05
sigtab.w = res.w[(res.w$padj < alpha), ]
sigtab.w = cbind(as(sigtab.w, "data.frame"), as(tax_table(physeq.whole)[rownames(sigtab.w), ], "matrix"))
sigtab.w
```

```r
# exclude SV that are not classifies at Genus level
sigtabgen.w = subset(sigtab.w, !is.na(Genus))
# Phylum order
x = tapply(sigtabgen.w$log2FoldChange, sigtabgen.w$Phylum, function(x) max(x))
x = sort(x, TRUE)
sigtabgen.w$Phylum = factor(as.character(sigtabgen.w$Phylum), levels=names(x))
# Genus order
x = tapply(sigtabgen.w$log2FoldChange, sigtabgen.w$Genus, function(x) max(x))
x = sort(x, TRUE)
sigtabgen.w$Genus = factor(as.character(sigtabgen.w$Genus), levels=names(x))
```

```r
# plot results
theme_set(theme_bw())
#theme_set(theme_light())
pdf('deseq2-16s-whole-raw.pdf')
ggplot(sigtabgen.w, aes(y=Genus, x=log2FoldChange, color=Phylum)) + 
  geom_vline(xintercept = 0.0, color = "gray", size = 0.5) +
  geom_point(size=6) + 
  theme(axis.text.x = element_text(angle = -90, hjust = 0, vjust=0.5))
dev.off()
```
