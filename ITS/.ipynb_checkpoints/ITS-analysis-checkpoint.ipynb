# Quantitative abundance

```r
# read in counts data
counts.df <- read.table("counts.txt", header=TRUE)
counts <- counts.df$sum
```

```r
counts
```

```r
# exclude R10 samples because they don't have counts data
# here physeq.23s.quan is not quantitative yet, later we will replace the current relative otu table with a new quantitative one
physeq.23s.rel <- subset_samples(physeq.28s.percent, SampleID != "T1R10" & SampleID != "T2R10" & SampleID != "T310" & SampleID != "T410" & SampleID != "T5R10")
```

```r
physeq.23s.quan <- physeq.23s.rel
nsamples(physeq.23s.quan)
```

```r
otu.table <- otu_table(physeq.23s.quan)
```

```r
# define a function to generate quantitative data by multiplying counts per sample
rel_to_quan <- function(physeq) {
    for (i in 1:23) {
        otu.table[,i] = get_taxa(physeq, sample_names(physeq)[i]) * counts[i] /100
    }
    return(otu.table)
}

# replace relative otu table with a new quantitative one by using rel_to_quan function
otu_table(physeq.23s.quan) <- rel_to_quan(physeq.23s.quan)
```

```r
# now physeq.23s.quan is officially a quantitative object
otu_table(physeq.23s.quan)
```

```r
colSums(otu_table(physeq.23s.quan))
```

```r
p.q <- plot_bar(physeq.23s.quan, x="SampleID", fill="Class") + theme_gray() + theme(axis.text.x = element_text(size = rel(0.7), angle = 45, hjust = 1))
p.q + geom_bar(stat = "identity") + ylab("Quantitative Abundance")
```

```r
colSums(counts(diagdds_rel))
```

```r
# calculate geometric means prior to estimate size factors
gm_mean = function(x, na.rm=TRUE){
  exp(sum(log(x[x > 0]), na.rm=na.rm) / length(x))
}
#geoMeans_rel = apply(counts(diagdds_rel), 1, gm_mean)
#diagdds_rel = estimateSizeFactors(diagdds_rel, geoMeans = geoMeans_rel)
#diagdds_rel = DESeq(diagdds_rel, fitType="local")
```

```r
res_rel <- results(diagdds_rel)
```

```r
summary(res_rel)
```

```r
res_rel_ordered = res_rel[order(res_rel$padj, na.last=NA), ]
```

```r
res_rel_ordered
```

```r
alpha = 0.05
# how many significant features
sum(res_rel$padj < alpha, na.rm=TRUE)

# what are they
sigtab_rel = res_rel_ordered[(res_rel_ordered$padj < alpha), ]
sigtab_rel
```

## Quantitative abundance data

```r
diagdds_quan = phyloseq_to_deseq2(physeq.23s.quan, ~ Material)
```

```r
#counts(diagdds_quan)
```

```r
geoMeans_quan = apply(counts(diagdds_quan), 1, gm_mean)
diagdds_quan = estimateSizeFactors(diagdds_quan, geoMeans = geoMeans_quan)
diagdds_quan = DESeq(diagdds_quan, fitType="local")
```

```r
res_quan <- results(diagdds_quan)
summary(res_quan)
```

```r
res_quan_ordered = res_quan[order(res_quan$padj, na.last=NA), ]
res_quan_ordered
```

```r
alpha = 0.05
# how many significant features
sum(res_quan$padj < alpha, na.rm=TRUE)

# what are they
sigtab_quan = res_quan_ordered[(res_quan_ordered$padj < alpha), ]
sigtab_quan
```

```r
sigtab_quan_df = cbind(as(sigtab_quan, "data.frame"), as(tax_table(physeq.23s.quan)[rownames(sigtab_quan), ], "matrix"))
sigtab_quan_df
```

```r
theme_set(theme_bw())
#sigtab_quan_df_gen = subset(sigtab_quan_df, !is.na(Genus))
sigtab_quan_df_gen = subset(sigtab_quan_df, Genus != "g__unidentified")
# Class order
x = tapply(sigtab_quan_df_gen$log2FoldChange, sigtab_quan_df_gen$Class, function(x) max(x))
x = sort(x, TRUE)
sigtab_quan_df_gen$Class = factor(as.character(sigtab_quan_df_gen$Class), levels=names(x))
# Genus order
x = tapply(sigtab_quan_df_gen$log2FoldChange, sigtab_quan_df_gen$Genus, function(x) max(x))
x = sort(x, TRUE)
sigtab_quan_df_gen$Genus = factor(as.character(sigtab_quan_df_gen$Genus), levels=names(x))

pdf("deseq2-ITS-23s-quan.pdf")
ggplot(sigtab_quan_df_gen, aes(y=Genus, x=log2FoldChange, color=Class)) + 
  geom_vline(xintercept = 0.0, color = "gray", size = 0.5) +
  geom_point(size=3) + 
  theme(axis.text.x = element_text(angle = -90, hjust = 0, vjust=0.5))
dev.off()
```

## All 28 samples

```r
physeq.28s
```

```r
otu_table(physeq.28s)
```

```r
diagdds = phyloseq_to_deseq2(physeq.28s, ~ Material)
```

```r
geoMeans = apply(counts(diagdds), 1, gm_mean)
diagdds = estimateSizeFactors(diagdds, geoMeans = geoMeans)
diagdds = DESeq(diagdds, fitType="local")
```

```r
res <- results(diagdds)
summary(res)
```

```r
res_ordered = res[order(res$padj, na.last=NA), ]
res_ordered
```

```r
alpha = 0.05
# how many significant features
sum(res$padj < alpha, na.rm=TRUE)

# what are they
sigtab = res_ordered[(res_ordered$padj < alpha), ]
sigtab
```

```r
sigtab_df = cbind(as(sigtab, "data.frame"), as(tax_table(physeq.28s)[rownames(sigtab), ], "matrix"))
sigtab_df
```

```r
theme_set(theme_bw())
#sigtab_df_gen = subset(sigtab_df, !is.na(Genus))
sigtab_df_gen = subset(sigtab_df, Genus != "g__unidentified")
# Class order
x = tapply(sigtab_df_gen$log2FoldChange, sigtab_df_gen$Class, function(x) max(x))
x = sort(x, TRUE)
sigtab_df_gen$Class = factor(as.character(sigtab_df_gen$Class), levels=names(x))
# Genus order
x = tapply(sigtab_df_gen$log2FoldChange, sigtab_df_gen$Genus, function(x) max(x))
x = sort(x, TRUE)
sigtab_df_gen$Genus = factor(as.character(sigtab_df_gen$Genus), levels=names(x))
           
pdf("deseq2-ITS-all-28s-raw.pdf")
ggplot(sigtab_df_gen, aes(y=Genus, x=log2FoldChange, color=Class)) + 
  geom_vline(xintercept = 0.0, color = "gray", size = 0.5) +
  geom_point(size=3) + 
  theme(axis.text.x = element_text(angle = -90, hjust = 0, vjust=0.5))
dev.off()
```
